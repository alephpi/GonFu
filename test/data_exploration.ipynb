```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = '../data/df_train.csv'
df = pd.read_csv(data_path, index_col=0)
```

```python
df.head()
```

```python
df.shape
```

```python
ax = [None] * 6
plt.figure(figsize= (20,10))
for i in range(1,7,1):
	ax[i-1] = plt.subplot(2,3,i)
	ax[i-1].hist(df.loc[df.index[0:df.shape[0]//2] , 's'+str(i)], alpha=0.5, bins=100)
	ax[i-1].hist(df.loc[df.index[df.shape[0]//2:] , 's'+str(i)], alpha=0.5, bins=100)
	# df.loc[df.index[0:df.shape[0]//2] , 's'+str(i)].hist(bins=100, ax=ax[i-1], legend=True)
	# df.loc[df.index[df.shape[0]//2:],'s'+str(i)].hist(bins=100, ax=ax[i-1], legend=True)
```
